### assignment 2_1

and 와 or은 논리연산자로 논리 연산을 따라 최종 결과가 변환되게 된다.
and 와 or 연산자의 연산 순서는 다음과 같다.
and는 왼쪽 값이 False이면 왼쪽 값을 반환, True이면 오른쪽 값의 연산을 따라서 진행하고 오른쪽 값도 True이면 오른쪽 값을 반환, False이면 왼쪽 값을 반환하고,
or는 왼쪽 값이 True이면 왼쪽 값을 반환, False이면 오른쪽 값을 반환한다.

위의 코드에서는 숫자값 사이의 논리 연산이 이루어졌는데 파이썬에서 0은 False, 그 외의 숫자는 True로 처리되므로 위의 연산 순서를 따라 연산되어 각 결과값이 출력되게 된다.


```python
def div_path(path):
    i = path.rfind('/')

    li = []
    li.append(path[:i])
    li.append(path[i+1:])
    
    return li

print div_path("/usr/local/bin/python")
```

```python
import urllib2
import re



        
source = urllib2.urlopen("http://cse.kut.ac.kr/").read()

#print removeTags(source)

```

```python
import string

def removeTags(html):
    rex =re.compile('<.*?>')
    text = re.sub(rex,'', html)
    
    return text

text = removeTags(source)

wordList = text.split()

removeWhiteSpace = list()
#리스트에서 공백문자 제거
for s in wordList:
    if s not in string.whitespace:
        removeWhiteSpace.append(s)

print removeWhiteSpace
```
