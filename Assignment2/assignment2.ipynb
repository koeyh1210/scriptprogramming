### assignment 2_1

and 와 or은 논리연산자로 논리 연산을 따라 최종 결과가 변환되게 된다.
and 와 or 연산자의 연산 순서는 다음과 같다.
and는 왼쪽 값이 False이면 왼쪽 값을 반환, True이면 오른쪽 값의 연산을 따라서 진행하고 오른쪽 값도 True이면 오른쪽 값을 반환, False이면 왼쪽 값을 반환하고,
or는 왼쪽 값이 True이면 왼쪽 값을 반환, False이면 오른쪽 값을 반환한다.

위의 코드에서는 숫자값 사이의 논리 연산이 이루어졌는데 파이썬에서 0은 False, 그 외의 숫자는 True로 처리되므로 위의 연산 순서를 따라 연산되어 각 결과값이 출력되게 된다.


```python
def div_path(path):
    i = path.rfind('/')

    li = []
    li.append(path[:i])
    li.append(path[i+1:])
    
    return li

print div_path("/usr/local/bin/python")
```

```python
import urllib2
import re



        
source = urllib2.urlopen("http://cse.kut.ac.kr/").read()

#print removeTags(source)

```

```python
import string

def removeTags(html):
    rex =re.compile('<.*?>')
    text = re.sub(rex,'', html)
    
    return text

text = removeTags(source)

wordList = text.split()

wordDic = {}

for word in wordList:
    if word not in wordDic:
        wordDic[word] = 1
    else :
        wordDic[word] += 1
    
for key in wordDic:
    print key, wordDic[key]
```

#### Incremental Project

정규표현식을 사용해 html태그를 제거 한 후 split()함수를 사용해 리스트자료형으로 변환 후 Dictionary 자료형을 사용하여 카운팅했습니다.

```python
maxPalNum = 0
palNum = 0

for i in range(100, 1000):
    for j in range(100, 1000):
        palNum = str(i * j)
        
        if palNum == palNum[::-1]:
            if int(palNum) > maxPalNum:
                maxPalNum = int(palNum)
                    
            
print maxPalNum
```

### 오일러 5번

n을 1~20까지의 수로 나누어 모두 나누어 떨어지면 그 수를 반환합니다.

```python
sum=0
pSum = 0
for i in range(1, 101):
    sum += i
    pSum += pow(i, 2)

sumPow = pow(sum, 2)

print "합의 제곱 : " + `sumPow`
print "제곱의 합 : " + `pSum`
print "두 수의 차 : " + `abs(sumPow - pSum)`
```
